```python
from shp2coco.tif_process import clip_from_file
from shp2coco.shape_to_coco import from_mask_to_coco
```

```python
from shp2coco.slice_dataset import slice
```

```python
import os
```

```python
# root path for saving the tif and shp file.
BASE_DIR = os.getcwd()
ROOT = os.path.join(BASE_DIR,'shp2coco/example_data/original_data')
# ROOT = r'shp2coco/example_data/original_data'
img_path = 'img'
shp_path = 'shp'
# root path for saving the mask.
ROOT_DIR = ROOT + '/dataset'
IMAGE_DIR = os.path.join(ROOT_DIR, "greenhouse_2019")
ANNOTATION_DIR = os.path.join(ROOT_DIR, "annotations")

clip_size = 300
```

```python
clip_from_file(clip_size, ROOT, img_path, shp_path)
slice(ROOT_DIR, train=0.7, eval=0.3, test=0)
from_mask_to_coco(ROOT_DIR, 'train', "greenhouse_2019", "annotations")
from_mask_to_coco(ROOT_DIR, 'eval', "greenhouse_2019", "annotations")
# from_mask_to_coco(ROOT_DIR, 'test', "greenhouse_2019", "annotations")
```
